```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('airquality_data.csv', low_memory=False , encoding='latin1')
```

```python
LINE CHART
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['date'][:50], df['so2'][:50])
plt.xticks(rotation=45)
plt.title("SO2 Line Chart")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=df['state'][:10], y=df['so2'][:10])
plt.xticks(rotation=45)
plt.title('SO2 Levels in Different States')
plt.show()
```

```python
plt.figure(figsize=(8,5))
correlation = df[['so2', 'no2', 'rspm', 'spm']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['so2'], y=df['no2'])
plt.title('SO2 vs NO2')
plt.xlabel('SO2')
plt.ylabel('NO2')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['so2'], bins=20)
plt.title('Distribution of SO2')
plt.xlabel('SO2')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.violinplot(y=df['no2'])
plt.title('Violin Plot of NO2')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(y=df['so2'])
plt.title('Boxplot of SO2')
plt.show()
```

```python

time_series = df.groupby('date')['so2'].mean()

plt.figure(figsize=(12,5))
plt.plot(time_series)
plt.title('Average SO2 Time Series')
plt.xlabel('Date')
plt.ylabel('Average SO2')

plt.xticks(rotation=45)

plt.show()
```
