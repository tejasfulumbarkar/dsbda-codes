```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
import numpy as np
import seaborn as sns 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PowerTransformer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score , mean_absolute_error , mean_squared_error
from sklearn.tree import DecisionTreeRegressor
```

```python
df = pd.read_csv('AQ_data.csv',low_memory=False )
df
```

```python
# Data cleaning 



# 1 remove nulls 

df.isnull().sum()
```

```python
#remove columns with too many nulls 
# first lets see column names and clean extra spaces 

df.columns = df.columns.str.strip()
# Drop columns with too many nulls (optional)
df.drop(['agency', 'location_monitoring_station'], axis=1, inplace=True)
```

```python
#fill missing values with mean

df.fillna(df.mean(numeric_only=True), inplace =True)
df
```

```python
#remove duplicate rows if any

df.drop_duplicates(inplace=True)
```

```python
df.dtypes
```

```python
df['stn_code'] = pd.to_numeric(df['stn_code'], errors='coerce')

df.dtypes
```

```python
df['stn_code'] = pd.to_numeric(df['stn_code'] , errors='coerce')
```

```python
#drop useless samppling column

df.drop(['sampling_date'] , inplace=True , errors='ignore' , axis=1)
df
```

```python
#data integration 


df['date'] = pd.to_datetime(df['date'] )
df.dtypes
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
#object bana liya 
scaler = StandardScaler()

#select numeriic columns

num_cols = ['so2','no2','rspm','spm','pm2_5']
```

```python
#apply normalization

df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
df
```

```python
# Convert categorical to numeric
df
```

```python
bool_cols = df.select_dtypes(include='bool').columns

df[bool_cols] = df[bool_cols].astype(int)
```

```python
df
```

```python
# model building 

#define x and y 

x = df[['so2','no2','rspm','spm']]
y = df['pm2_5']
```

```python
x 
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size =0.2)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print("R2 Score:", r2_score(y_test , y_pred))
```

```python
print ("MAE:" , mean_absolute_error(y_test,y_pred))
```

```python
model_1 = DecisionTreeRegressor()
model_1.fit(x_train, y_train)

y_pred = model_1.predict(x_test)

print("R2:", r2_score(y_test, y_pred))
```
