```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
import seaborn as sns
import os 
```

```python
os.getcwd()
df = pd.read_csv("heart.csv")
```

```python
df.head()
```

```python
#line chart
df_sorted = df.sort_values(by='age')
plt.figure()
plt.plot(df_sorted['age'],df_sorted['chol'],marker='.')
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.title("Line Chart: Age vs Cholesterol")
plt.show()
```

```python
plt.figure()
df['sex'].value_counts().plot(kind='bar')
plt.xlabel("Sex")
plt.ylabel("Count")
plt.title("Bar Plot: Sex Distribution")
plt.show()
```

```python
plt.figure()
df['sex'].value_counts().plot(kind='bar')
plt.xlabel("Sex")
plt.ylabel("Count")
plt.title("Bar Plot: Sex Distribution")
plt.show()
```

```python
plt.figure(figsize=(12,8))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm",)
plt.title("Correlation Heatmap")
plt.show()

# corr calculates correlation berween two variable 
# corr shows how strongly two varibles are related 
```

```python
plt.figure()
plt.scatter(df['age'], df['thalachh'])
plt.xlabel("Age")
plt.ylabel("Max Heart Rate")
plt.title("Scatter Plot: Age vs Thalach")
plt.show()
```

```python
plt.figure()
plt.hist(df['chol'], bins=20)
plt.xlabel("Cholesterol")
plt.title("Histogram of Cholesterol")
plt.show()
```

```python
plt.figure()
plt.boxplot(df['chol'])
plt.title("Box Plot of Cholesterol")
plt.show()
```

```python
plt.figure()
sns.violinplot(y=df['chol'])
plt.title("Violin Plot of Cholesterol")
plt.show()
```

```python
df['time'] = range(len(df))

plt.figure()
plt.plot(df['time'], df['chol'])
plt.xlabel("Time")
plt.ylabel("Cholesterol")
plt.title("Time Series Plot")
plt.show()
```
